# Registration and Login system using Python, file handling#

```python
import re

email_regex = re.compile(r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')
pwd_regex = re.compile(r'^(?=.*?[A-Z])(?=.*?[a-z])(?=.*?[0-9])(?=.*?[#?!@$%^&*-]).{5,}$')

def V_mail(email):
    if re.fullmatch(email_regex, email):
      return True
    else:
      return False

def V_pwd(pwd):
    if len(pwd) < 16 and len(pwd) > 5 and re.fullmatch(pwd_regex, pwd):
      return True
    else:
      return False


def register():
    db = open ("databases.txt", "r")
    email = input("create email: ")
    d =[]
    f =[]
    for i in db:
        a,b = i.split(", ")
        b = b.strip()
        d.append(a)
        f.append(b)
    data = dict(zip(d,f))
 
    if V_mail(email):
        pwd = input("create password: ")
        if V_pwd(pwd):
            db = open("databases.txt", "a")
            db.write(email+", "+pwd+"\n")
            print('User Registered Successfully')
        else:
            print('Invalid Password, Please Try Again...')
    else:
        print('Invalid Username, Please Try Again...')
    
def access():
    db = open("databases.txt", "r")
    email = input("Enter your email: ")
    pwd = input("Enter your password: ")
    
    if not len(email or pwd)<1:
        d =[]
        f =[]
        for i in db:
            a,b = i.split(", ")
            b = b.strip()
            d.append(a)
            f.append(b)
        data = dict(zip(d,f))
        
        try:
            if data[email]:
                try:
                    if pwd == data[email]:
                        print("Login Success")
                        print("Hi, Welcome")
                    else:
                        print("Password or Username incorrect")
                except:
                    print("Incorrect password or username")
            else:
                print("*Please Register!!!")
        except:
            print("*Please Register!!!")
    else:
        print("Please enter a value")

def home(option=None):
    option = input("Login | signup: ")
    if option == "Login":
        access()
    elif option =="signup":
        register()
    else:
        print("Please enter an option")
home()
```
